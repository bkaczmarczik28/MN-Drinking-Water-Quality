```python
#Create path to csv for arsenic 2017 data and read into Pandas Dataframe
csv_path = "arsenic_2017.csv"
arsenic2017_df= pd.read_csv(csv_path)

arsenic2017_df.head()
```

```python
#verify data types in pandas database for arsenic
arsenic2017_df.dtypes
```

```python
#delete the empty column on the end named "Unnamed:11"
arsenic2017_df=arsenic2017_df.drop(columns=['Unnamed: 11'])
arsenic2017_df.head()
```

```python
#verify any empty data cells
arsenic2017_df['mean'].isnull().sum()
```

```python
#verify what is in the empty data cells
nullData=arsenic2017_df.loc[arsenic2017_df['mean'].isnull()]
nullData
```

```python
#delete the null data and confirm no more NaN
arsenic2017_df=arsenic2017_df.dropna()
arsenic2017_df['mean'].isnull().sum()
```

```python
#drop concentration string out of contaminate level columns "mean" and "max"
arsenic2017_df['mean']=arsenic2017_df['mean'].str.split(" ").str[0]
arsenic2017_df['max']=arsenic2017_df['max'].str.split(" ").str[0]
arsenic2017_df.head()
```

```python
#rename column headers for contaminates to include concentrations so the information isn't lost
arsenic2017_df=arsenic2017_df.rename(columns={"mean":"Mean ug/L", "max":"Max ug/L"})
arsenic2017_df.head()
```

```python
#convert contaminate columns from object to integer, confirm data type in columns
arsenic2017_df["Mean ug/L"] = pd.to_numeric(arsenic2017_df["Mean ug/L"])
arsenic2017_df["Max ug/L"] = pd.to_numeric(arsenic2017_df["Max ug/L"])
print(arsenic2017_df.dtypes)
```
